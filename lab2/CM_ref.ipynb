# Data label processing

```python
import torch
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor
import cv2
from ultralytics import YOLO
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

```python
os.makedirs('/content/train/mask/', exist_ok=True)
os.makedirs('/content/train/annotation/', exist_ok=True)
for f in tqdm(os.listdir('/content/train/training_set')):
    if 'Annotation' in f:
        mask = cv2.imread(f'/content/train/training_set/{f}', 0)
        coords = np.where(mask == 255)

        # Find convex hull of those coords
        hull = np.squeeze(cv2.convexHull(np.array(coords).T))


        string = '0'
        for x, y in hull:
            string += f' {y/mask.shape[1]} {x/mask.shape[0]}'

        # Write label
        with open(f'/content/train/annotation/{f[:-15]}.txt', 'w') as label:
            label.write(string)

        os.rename(f'/content/train/training_set/{f}', f'/content/train/mask/{f}')
```

## Images

```python
for image_name in os.listdir('/content/test/test_set')[:10]:
    img = cv2.imread(f'/content/test/test_set/{image_name}')
    results = model.predict(img)[0]

    color_dict = {
        0: (0, 0, 255)
    }

    # Draw bbox
    for cat, score, mask in zip(results.boxes.cls, results.boxes.conf, results.masks.xy):
        mask = mask.reshape(-1, 1, 2)
        hull = cv2.convexHull(mask)
        ellipse = cv2.fitEllipse(hull)
        (center, axes, angle) = ellipse
        grain_length = max(axes)
        grain_width = min(axes)


        # Calculate the endpoints of the major axis
        major_axis_length = grain_length / 2
        major_axis_vector = np.array([np.cos(np.radians(angle+90)), np.sin(np.radians(angle+90))])
        major_axis_endpoint1 = np.intp(center + major_axis_length * major_axis_vector)
        major_axis_endpoint2 = np.intp(center - major_axis_length * major_axis_vector)

        cv2.line(img, tuple(major_axis_endpoint1), tuple(major_axis_endpoint2), (255, 0, 0), 2)


        # Calculate the start and end points for the line perpendicular to the grain length
        line_length = grain_width  # Length of the line perpendicular to the grain length
        dx = int(line_length / 2 * np.cos(np.deg2rad(angle)))
        dy = int(line_length / 2 * np.sin(np.deg2rad(angle)))
        start_point = (int(center[0]) - dx, int(center[1]) - dy)
        end_point = (int(center[0]) + dx, int(center[1]) + dy)

        # Draw grain width
        cv2.line(img, start_point, end_point, (0, 0, 255), 2)

        # Draw the ellipse
        cv2.ellipse(img, ellipse, (255, 255, 0), 2)

        # cv2.polylines(img, [mask], color_dict[cat.item()], 2)
        # print(cat, score)
        # cv2.putText(img, f'{results.names[cat.item()]}', (mask[0], mask[1] + 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

    plt.imshow(img)
    plt.show()
```

# Measure


```python
def peri_elli(minor, major):
    return np.pi * (minor + major) * (1 + (3 * ((minor - major) / (minor + major)) ** 2) / (10 + np.sqrt(4 - 3 * ((minor - major) / (minor + major)) ** 2)))
```

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import mean_squared_error as mse
import pandas as pd
```

```python
df = pd.read_csv('/content/training_set_pixel_size_and_HC.csv')
```

```python
from tqdm import tqdm
pred_peri = []
actual_peri = []

for image_name in tqdm(sorted(os.listdir('/content/train/training_set'))):
    img = cv2.imread(f'/content/train/training_set/{image_name}')
    results = model.predict(img, verbose=False)[0]

    if results.masks is None:
        continue

    for cat, score, mask in zip(results.boxes.cls, results.boxes.conf, results.masks.xy):
        mask = mask.reshape(-1, 1, 2)
        hull = cv2.convexHull(mask)
        ellipse = cv2.fitEllipse(hull)
        (center, axes, angle) = ellipse
        major = max(axes)
        minor = min(axes)

    pred_peri.append(peri_elli(minor, major)*df[df['filename'] == image_name]['pixel size(mm)'].iloc[0])
    actual_peri.append(df[df['filename'] == image_name]['head circumference (mm)'].iloc[0])






```

```python
print(f'mAPE of peri is {mape(actual_peri, pred_peri)}')
```

```python
print(f'RMSE of peri is {np.sqrt(mse(actual_peri, pred_peri))}mm')
```
